```python
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

Let's Upload the IRIS data

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 3, min_samples_leaf = 1)
tree_clf.fit(X, y)
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3)
leaves = range(1, 11)
scores =[]
for leaf in leaves:
  clf = tree.DecisionTreeClassifier(random_state = 2022, max_depth = 3, min_samples_leaf = leaf)
  score = cross_val_score(clf, X_train, y_train, cv = 3)
  scores.append(score.mean())
plt.plot(leaves,scores)
plt.xlabel("min_sample_leaf")
plt.ylabel("Cross Validation Accuracy")
plt.title("Cross Validation Score vs min_sample_leaf for Decision Tree")
plt.show()
```

```python
clf = tree.DecisionTreeClassifier(random_state = 2022, max_depth = 3, min_samples_leaf = 3)
estimators = range(1, 201)
scores =[]
for estimator in estimators:
  clf = GradientBoostingClassifier(random_state = 2022, max_depth = 2, n_estimators = estimator)
  score = cross_val_score(clf, X_train, y_train, cv = 3)
  scores.append(score.mean())

plt.plot(estimators, scores)
plt.xlabel("min_sample_leaf")
plt.ylabel("Cross Validation Accuracy")
plt.title("Cross Validation Score vs min_sample_leaf for Gradient Boosting Decision Tree")
plt.show()
```
